###2. Data Description

The Data is collected from mobile apps that use InsaidTelecom services. 

The data schema can be represented in the following table:

**gender_age_train** - Devices and their respective user gender, age and age_group

**phone_brand_device_model** - device ids, brand, and models phone_brand ( few brands are in Chinese)

**events_data** - when a user uses mobile on INSAID Telecom network, the event gets logged in this data.
Each event has an event id, location (lat/long), and the event corresponds to frequency of mobile usage. timestamp: when the user is using the mobile.
